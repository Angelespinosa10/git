```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Subir imagen desde el equipo local
uploaded = files.upload()

# Leer la imagen subida
image_path = list(uploaded.keys())[0]  # Nombre del archivo subido
imagen = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Mostrar la imagen cargada
plt.imshow(imagen, cmap='gray')
plt.title("Imagen Original")
plt.show()

# Aplicar un umbral para convertir la imagen a binaria (0 para fondo, 255 para la mancha)
_, imagen_binaria = cv2.threshold(imagen, 127, 255, cv2.THRESH_BINARY)

# Mostrar la imagen binaria resultante
plt.imshow(imagen_binaria, cmap='gray')
plt.title("Imagen Binaria")
plt.show()

# Calcular el área de la mancha contando los píxeles de valor 255
area_manchada = np.sum(imagen_binaria == 255)

# Imprimir el área calculada
print(f"Área de la mancha: {area_manchada} píxeles")
```
